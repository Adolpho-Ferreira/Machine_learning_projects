# Levantar as métricas

```python
#Renomeando Colunas para trabalhar mais fácil
organics = sheet3
organics = organics.rename(columns={'Acquisition Channel': 'Acquisition',
                                    'Store Listing Visitors': 'Vis',
                                    'Visitor-to-Installer conversion rate': 'Vis_inst_rate',
                                    'Installers retained for 1 day':'Inst_ret_1',
                                    'Installer-to-1 day retention rate': 'Inst_ret_rate_1',
                                    'Installers retained for 7 days': 'Inst_ret_7',
                                    'Installer-to-7 days retention rate': 'Inst_ret_rate_7',
                                    'Installers retained for 15 days': 'Inst_ret_15',
                                    'Installer-to-15 days retention rate': 'Inst_ret_rate_15',
                                    'Installers retained for 30 days': 'Inst_ret_30',
                                    'Installer-to-30 days retention rate': 'Inst_ret_rate_30'}, inplace = False)
organics['Acquisition'][organics['Acquisition'].str.contains('channels')] = 'Tracked'
```

```python
#Filtrando Apenas Métricas Orgnânicas
organics_index = (organics['Acquisition'] == 'Organic')
organics = organics[organics_index]
```

```python
#Número de Instalações Orgânicas
organics_downloads = organics.Installers.sum()
```

```python
#Número de Visitantes Orgânicos
organics_visitors = organics.Vis.sum()
```

```python
#Calculando Médias Orgânicas
organics_conv = np.mean(organics.Vis_inst_rate)
organics_ret_1 = np.mean(organics.Inst_ret_1)
organics_ret_rate_1 = np.mean(organics.Inst_ret_rate_1)
organics_ret_7 = np.mean(organics.Inst_ret_7)
organics_ret_rate_7 = np.mean(organics.Inst_ret_rate_7)
organics_ret_15 = np.mean(organics.Inst_ret_15)
organics_ret_rate_15 = np.mean(organics.Inst_ret_rate_15)
organics_ret_30 = np.mean(organics.Inst_ret_30)
organics_ret_rate_30 = np.mean(organics.Inst_ret_rate_30)
```

```python
#Separando Motivos Negativos e Positivos em Duas Tabelas
motive = sheet2
motive_neg_index = (motive['Sentiment'] == 'Negative')
motive_pos_index = (motive['Sentiment'] == 'Positive')
motive_neg = motive[motive_neg_index]
motive_pos = motive[motive_pos_index]
```

```python
#Diminuindo Tabela Para Trabalhar mais Rápido
to_drop = ['Package Name', 'App Version Name', 'Reviewer Language', 'Device', 'Review Last Update Date and Time', 'Star Rating', 'Review Title']
motive_neg.drop(to_drop, inplace=True, axis=1)
motive_pos.drop(to_drop, inplace=True, axis=1)
```

```python
#Medindo Reviews Totais, Positivos e Negativos
motive_total = len(motive)
neg_rev_num = len(motive_neg)
pos_rev_num = len(motive_pos)
```

```python
#Filtrando Se Possui Texto e Calculando Quantos Sobram
motive_neg = motive_neg['Review Text'].dropna()
neg_rev_comtxt_num = len(motive_neg)
motive_pos = motive_pos['Review Text'].dropna()
pos_rev_comtxt_num = len(motive_pos)
```

```python
#Calcular o churn
organics_churn_rate_1 = (1 -organics_ret_rate_1)
organics_churn_rate_7 = (1 -organics_ret_rate_7)
organics_churn_rate_15 = (1 -organics_ret_rate_15)
organics_churn_rate_30 = (1 -organics_ret_rate_30)
```

```python
#Calcular Estimativa Quantos Foram Capturados e Permanecerão por Mais de 30 Dias
organics_cres = (organics_downloads * organics_ret_rate_30)
organics_cres = int(organics_cres)
```

```python
#Renomeando para Precaver Danificação do Original
s2 = sheet2
s2 = s2.rename(columns={'Star Rating': 'Rating'}, inplace = False)
```

```python
#Calcular média das avaliações
ratings = np.mean(s2.Rating)
```

```python
#Calcular o organic Uplift 1/2
#Já foi calculado instalações organicas em organics_downloads
s3 = sheet3
s3 = s3.rename(columns={'Acquisition Channel': 'Acquisition'}, inplace = False)
non_organics_index = (s3['Acquisition'] != 'Organic')
non_organics = s3[non_organics_index]
```

```python
#Calcular o organic Uplift 2/2
non_organics_downloads = non_organics.Installers.sum()
organic_uplift = (organics_downloads / non_organics_downloads)
```

```python
#Criando Tabela das Folhas 1/3
data = {'Organic Store Visitors':  [organics_visitors],
        'Installers': [organics_downloads],
        'Visitors-to-Installers Conversion Rate': [organics_conv],
        'Installer retained for 1 day': [organics_ret_1],
        'Installer-to-1 day retained conversion': [organics_ret_rate_1],
        'Installer retained for 7 days': [organics_ret_7],
        'Installer-to-7 days retained conversion': [organics_ret_rate_7],
        'Installer retained for 15 days': [organics_ret_15],
        'Installer-to-15 days retained conversion': [organics_ret_rate_15],
        'Installer retained for 30 days': [organics_ret_30],
        'Installer-to-30 days retained conversion': [organics_ret_rate_30]}
out_1 = pd.DataFrame (data, columns = ['Organic Store Visitors',
                                       'Installers',
                                       'Visitors-to-Installers Conversion Rate',
                                       'Installer retained for 1 day',
                                       'Installer-to-1 day retained conversion',
                                       'Installer retained for 7 days',
                                       'Installer-to-7 days retained conversion',
                                       'Installer retained for 15 days',
                                       'Installer-to-15 days retained conversion',
                                       'Installer retained for 30 days',
                                       'Installer-to-30 days retained conversion'])
```

```python
#Criando Tabela das Folhas 2/3
data = {'Churn Rate 1 day':  [organics_churn_rate_1],
        'Churn Rate 7 days': [organics_churn_rate_7],
        'Churn Rate 15 days': [organics_churn_rate_15],
        'Churn Rate 30 days': [organics_churn_rate_30],
        'Installers retained for at least 30 days': [organics_cres],
        'Organic growing rate': [organics_conv],
        'Organic uplifting': [organic_uplift]}
out_2 = pd.DataFrame (data, columns = ['Churn Rate 1 day',
                                       'Churn Rate 7 days',
                                       'Churn Rate 15 days',
                                       'Churn Rate 30 days',
                                       'Installers retained for at least 30 days',
                                       'Organic growing rate',
                                       'Organic uplifting'])
```

```python
#Criando Tabela das Folhas 3/3
data = {'Rating': [ratings],
        'Number of reviews': [motive_total],
        'Reviews with positive sentiment': [pos_rev_num],
        'Reviews with positive sentiment and write a message': [pos_rev_comtxt_num],
        'Reviews with negative sentiment': [neg_rev_num],
        'Reviews with negative sentiment and write a message': [neg_rev_comtxt_num]}
out_3 = pd.DataFrame (data, columns = ['Rating',
        'Number of reviews',
        'Reviews with positive sentiment',
        'Reviews with positive sentiment and write a message',
        'Reviews with negative sentiment',
        'Reviews with negative sentiment and write a message'])
```
